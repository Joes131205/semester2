```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
```

![Formula](image.png)

```python
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [2, 5, 1, 6, 2, 3, 10, 7]

x = np.array(x)
y = np.array(y)

A = np.vstack([x, np.ones(len(x))]).T

y = y[:, np.newaxis]

alpha = np.dot(np.dot(np.linalg.inv(np.dot(A.T, A)), A.T), y)

# alpha[0] = slope, alpha[1] = intercept

# y = mx + b

plt.plot(x, y, 'bo')
plt.plot(x, alpha[0] * x + alpha[1], 'r')
plt.show()
```
